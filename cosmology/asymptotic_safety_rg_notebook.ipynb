# Asymptotic Safety – Illustrative Renormalization Group Notebook

This notebook accompanies a theoretical essay on asymptotically safe quantum gravity.
It provides **toy numerical and visualization experiments** illustrating key ideas:

- Renormalization group flows
- UV fixed points
- Relevant vs irrelevant couplings
- Weak‑gravity bound behavior
- RG trajectories connecting UV → IR

The models used here are **simplified toy beta functions** meant to illustrate structure rather than reproduce full FRG computations.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

## Toy RG System

We define simplified beta functions for three couplings:

- `g` : dimensionless Newton coupling
- `lam` : cosmological constant
- `h` : Horndeski braiding coupling

These capture qualitative RG behavior discussed in asymptotic safety literature.

```python
def beta(t, y):
    g, lam, h = y
    dg = 2*g - g**2
    dlam = -2*lam + g
    dh = -h*(1+g)
    return [dg, dlam, dh]
```

## Integrating an RG Trajectory

```python
t_span = [0, -10]
y0 = [0.5, 0.1, 0.5]

sol = solve_ivp(beta, t_span, y0, dense_output=True)

t = np.linspace(0, -10, 400)
g, lam, h = sol.sol(t)
```

## Coupling Evolution

```python
plt.plot(t, g, label='g')
plt.plot(t, lam, label='lambda')
plt.plot(t, h, label='h')
plt.xlabel('RG time')
plt.ylabel('Coupling value')
plt.legend()
plt.show()
```

## RG Phase Diagram

```python
g_vals = np.linspace(0,3,20)
l_vals = np.linspace(-1,1,20)

G, L = np.meshgrid(g_vals,l_vals)

dg = 2*G - G**2
dl = -2*L + G

plt.streamplot(G,L,dg,dl)
plt.xlabel('g')
plt.ylabel('lambda')
plt.show()
```

## Weak Gravity Bound Illustration

Toy matter coupling:

β_y = y (1 − y − g)

For sufficiently strong gravity, the matter fixed point disappears.

```python
g_range = np.linspace(0,2,200)
y_star = 1 - g_range

plt.plot(g_range, y_star)
plt.axhline(0)
plt.axvline(1, linestyle='--')
plt.xlabel('g')
plt.ylabel('y*')
plt.show()
```

## RG‑Improved Newton Constant

```python
k = np.logspace(-4,2,400)
G0 = 1
Gk = G0/(1 + G0*k**2)

plt.loglog(k, Gk)
plt.xlabel('scale k')
plt.ylabel('G(k)')
plt.show()
```

## Stability Matrix and Critical Exponents

We compute the stability matrix at the fixed point.

```python
import sympy as sp

g, lam, h = sp.symbols('g lam h')

beta_g = 2*g - g**2
beta_l = -2*lam + g
beta_h = -h*(1+g)

B = sp.Matrix([
    [sp.diff(beta_g,g), sp.diff(beta_g,lam), sp.diff(beta_g,h)],
    [sp.diff(beta_l,g), sp.diff(beta_l,lam), sp.diff(beta_l,h)],
    [sp.diff(beta_h,g), sp.diff(beta_h,lam), sp.diff(beta_h,h)]
])

fixed_point = {g:2, lam:1, h:0}

B_fp = B.subs(fixed_point)
B_fp.eigenvals()
```
