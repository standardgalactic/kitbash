```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
import sympy as sp

plt.rcParams["figure.figsize"] = (8,6)
```

## 1. Renormalization Group Intuition

The renormalization group describes how the parameters of a physical theory change when we change the **energy scale** at which we observe the system.

Instead of a single fixed theory, we obtain a **flow of couplings**.

Mathematically this looks like

\[
k \frac{d g_i}{dk} = \beta_i(g)
\]

where:

- `g_i` are couplings
- `β_i` are beta functions
- `k` is the energy scale

```python
def beta_simple(t, g):
    return 2*g - g**2

t = np.linspace(-5,5,200)

g = solve_ivp(lambda t,y: beta_simple(t,y), [0,-5], [0.2], dense_output=True).sol(t)[0]

plt.plot(t,g)
plt.xlabel("RG time")
plt.ylabel("coupling g")
plt.title("Example RG flow")
plt.show()
```

## 2. Fixed Points

A **fixed point** occurs when the beta function vanishes

\[
\beta(g_*) = 0
\]

At a fixed point the theory becomes scale invariant.

```python
g = sp.symbols('g')

beta = 2*g - g**2

sp.solve(beta, g)
```

These fixed points are

- `g = 0` (Gaussian)
- `g = 2` (interacting)

## 3. Stability and Critical Exponents

To understand behavior near a fixed point we compute

\[
B = \frac{d\beta}{dg}
\]

```python
sp.diff(beta, g)
```

```python
float(sp.diff(beta,g).subs(g,2))
```

If the derivative is negative the direction is **irrelevant**.

## 4. Relevant vs Irrelevant Couplings

Relevant couplings grow as we move away from the UV fixed point.

Irrelevant couplings shrink.

This determines the **predictive power of the theory**.

```python
def beta_two(t,y):
    g, h = y

    dg = 2*g - g**2
    dh = -h*(1+g)

    return [dg, dh]

sol = solve_ivp(beta_two, [0,-10], [0.5,0.8], dense_output=True)

t = np.linspace(0,-10,300)

g,h = sol.sol(t)

plt.plot(t,g,label="relevant g")
plt.plot(t,h,label="irrelevant h")

plt.legend()
plt.show()
```

The coupling `h` flows to zero automatically.

This is exactly the phenomenon responsible for the **suppression of the Horndeski braiding coupling** in the essay.

## 5. Universality

Different microscopic theories can flow to the **same fixed point**.

This is why critical phenomena in condensed matter show universal behavior.

```python
initial_conditions = [0.2,0.5,1.2]

for g0 in initial_conditions:

    sol = solve_ivp(lambda t,y: beta_simple(t,y), [0,-5], [g0], dense_output=True)

    g = sol.sol(t)[0]

    plt.plot(t,g)

plt.title("Different starting points approaching same structure")
plt.show()
```

## 6. Phase Diagrams in Theory Space

```python
g_vals = np.linspace(-1,3,20)
h_vals = np.linspace(-1,1,20)

G,H = np.meshgrid(g_vals,h_vals)

dG = 2*G - G**2
dH = -H*(1+G)

plt.streamplot(G,H,dG,dH)
plt.xlabel("g")
plt.ylabel("h")
plt.title("RG phase diagram")
plt.show()
```

## 7. Effective Field Theory

Effective field theories describe physics below some cutoff scale.

They are valid approximations even if the fundamental theory is unknown.

However a fundamental theory must remain **consistent at arbitrarily high energies**.

## 8. Asymptotic Safety Idea

A theory is asymptotically safe if

- an interacting UV fixed point exists
- only a finite number of relevant directions appear

## 9. Weak Gravity Bound Intuition

Strong gravitational fluctuations can destroy matter fixed points.

```python
g_range = np.linspace(0,2,200)

y_star = 1-g_range

plt.plot(g_range,y_star)
plt.axvline(1,color="red",linestyle="--")

plt.xlabel("gravity strength g")
plt.ylabel("matter fixed point")

plt.title("Toy weak‑gravity bound")
plt.show()
```
